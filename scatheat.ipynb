```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df,hue="species");
```

```python
sns.pairplot(df,hue="species",markers=["o","s","D"]);
```

```python
sns.pairplot(df,kind="reg");
```

```python
sns.pairplot(df,hue="species",kind="reg");
```

# heatmap

```python
flights=sns.load_dataset('flights')
df=flights.copy()
df.head()
```

```python
#uzun vadeli veriler(yıl ve ay),sayısal değerler,periyodik olaylarda işe yarar
```

```python
df.shape
```

```python
df["passengers"].describe()
```

```python
df=df.pivot("month","year","passengers");
```

```python
df
```

```python
#yapısal sorunsal
sns.heatmap(df);
```

```python
sns.heatmap(df,annot=True,fmt="d");
```

```python
sns.heatmap(df,annot=True,fmt="d"),linewidths=.5);
```

```python
sns.heatmap(df,annot=True,fmt="d",linewidths=.5);
```

# line plot

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",
             style="event",markers=True,dashes=False,data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="region",
             style="event",markers=True,dashes=False,data=df);
```
